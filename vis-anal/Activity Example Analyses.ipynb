```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
income = pd.read_csv('income.csv')
```

```python
income.head()
```

```python
sns.relplot(x="year", y="income", hue="region", kind="line", ci=None, data=income)
```

```python
sns.relplot(x="year", y="income", col="region", kind="line", ci=None, data=income)
```

```python
sns.relplot(x="year", y="income", col="country", col_wrap=5, kind="line", ci=None, data=income[income.region=="Europe & Central Asia"])
```
